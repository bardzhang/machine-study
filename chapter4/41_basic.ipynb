```python
import torch
import matplotlib.pyplot as plt

x=torch.arange(-8.0,8.0,0.1,requires_grad=True)
y=torch.relu(x)          #relu=max(x,0)   prelu=max(0,x)+a*min(0,x)

plt.plot(x.detach(),y.detach())
plt.xlabel("x")
plt.xlabel("y")ÃŸ
plt.grid()
plt.show() 
```

```python
y.backward(torch.ones_like(x))

plt.plot(x.detach(),x.grad)
plt.xlabel("x")
plt.ylabel("grad x")
plt.grid()
plt.show()
```

```python
y=torch.sigmoid(x)
plt.plot(x.detach(),y.detach())
plt.grid()
plt.show()
```

```python
x.grad.zero_()
y.backward(torch.ones_like(x))
plt.plot(x.detach(),x.grad)
plt.grid()
plt.show()
```

```python
y=torch.tanh(x)
plt.plot(x.detach(),y.detach())
plt.grid()
plt.show()
```

```python
x.grad.data.zero_()
y.backward(torch.ones_like(x),retain_graph=True)
plt.plot(x.detach(),x.grad)
plt.grid()
plt.show()
```
