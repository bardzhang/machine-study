```python
import torch
```

```python
x=torch.arange(4,dtype=torch.float32)
x.requires_grad_(True)
x.grad
```

```python
y=2*torch.dot(x,x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad.zero_()
y=x.sum()    #y=x1+x2
y.backward()
x.grad
```

```python
x=torch.tensor([1.0,2.0],requires_grad=True)
y=x**2         #y=[x1^2,x2^2]  
y.backward(torch.tensor([1.0,1.0]))  #v*jacobian 将y对x的雅可比矩阵乘以v
x.grad
```

```python
def f(x):
    return x**2

x=torch.tensor([1.0,2.0],requires_grad=True)
y=f(x)
j=torch.autograd.functional.jacobian(f,x)
print(j)
```
